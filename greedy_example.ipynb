```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import numpy as np

plt.style.use("seaborn-v0_8-muted")
```

```python
from spaces import Box
from optimization_problems import QuadraticOptimization
```

```python
X = Box([0, 0], [20, 20], dtype=np.int32)
```

```python
from greedy_algorithms import simpleGreedy, powerGreedy
```

```python
opt = QuadraticOptimization([-1, -1], [40, 50], -349, lambda x: x[0] + x[1])
```

```python
def drawPlots(plots, name):
    MXX = 16
    MXY = 12
    fig, ax = plt.subplots(figsize=(MXX / 1.5, MXY / 1.5))
    ax.add_patch(Circle([20, 25], 26, color="#aaaaaa"))
    plt.xlim(-0.5, MXX)
    plt.ylim(-0.5, MXY)
    ax.set_xticks(np.arange(0, MXX + 1, 1))
    ax.set_yticks(np.arange(0, MXY + 1, 1))
    ax.grid(which="major", color="k")
    for x, y, label in plots:
        plt.plot(x, y, marker="o", markersize=10, linewidth=3, label=label)
    plt.legend()
    plt.savefig(f"data/pictures/{name}.png")
    plt.show()
```

```python
def getPathXY(path):
    return [x[0] for x in path], [x[1] for x in path]

sg, sg_path = simpleGreedy(opt, X, [1, 2], render_path=True)
sgx, sgy = getPathXY(sg_path)

pg, pg_path = powerGreedy(opt, X, [1, 2], render_path=True)
pgx, pgy = getPathXY(pg_path)
```

```python
drawPlots([(sgx, sgy, "Simple Greedy")], "only_simple_greedy_quadratic")
```

```python
drawPlots([(sgx, sgy, "Simple Greedy"), (pgx, pgy, "Power Greedy")], "simple_and_power_greedy_quadratic")
```
